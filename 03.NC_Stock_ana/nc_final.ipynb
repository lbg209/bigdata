```python
import pandas as pd
import matplotlib.pyplot as plt
from functools import reduce
%matplotlib inline
```

```python
nc_df = pd.read_csv('/gdrive/MyDrive/Colab Notebooks/NC_stock/csv/nc_stock_data.csv')
blade_df = pd.read_csv('/gdrive/MyDrive/Colab Notebooks/NC_stock/csv/blade_data.csv')
osstem_df = pd.read_csv('/gdrive/MyDrive/Colab Notebooks/NC_stock/csv/osstem_data.csv')
nft_df = pd.read_csv('/gdrive/MyDrive/Colab Notebooks/NC_stock/csv/nft_data.csv')
lineage_df = pd.read_csv('/gdrive/MyDrive/Colab Notebooks/NC_stock/csv/lineage_data.csv')
```

```python
blade_df.rename(columns={'num_search':'blade_search'},inplace=True)
osstem_df.rename(columns={'num_search':'osstem_search'},inplace=True)
nft_df.rename(columns={'num_search':'nft_search'},inplace=True)
lineage_df.rename(columns={'num_search':'lineage_search'},inplace=True)
```

```python
dfs = [nc_df, blade_df, osstem_df,lineage_df, nft_df]
df_merge = reduce(lambda left, right: pd.merge(left, right, on='date'), dfs)

df_merge
```

키워드 : 블레이드앤소울, 리니지

```python
fig = plt.figure(figsize=(30, 6))
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()

x_date = nc_df['date'].to_list()
y_search = nc_df['num_search'].to_list()
y_stock = nc_df['stock_price'].to_list()
y_blade = blade_df['blade_search'].to_list()
y_osstem = osstem_df['osstem_search'].to_list()
y_nft = nft_df['nft_search'].to_list()
y_lineage = lineage_df['lineage_search'].to_list()

ax1.bar(x_date, y_stock, width=0.5, color='grey')
ax2.plot(x_date, y_blade, marker='o', color = 'red')
ax2.plot(x_date, y_lineage, marker='o', color = 'pink')

ax1.set_xlabel('date')
ax1.set_xticklabels(x_date, rotation = 90)
ax1.set_ylabel('stock price')
ax2.set_ylabel('search')
```

키워드 : NFT, 오스템

```python
fig = plt.figure(figsize=(30, 6))

ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()

ax1.bar(x_date,y_stock,width=0.5,color='grey')
ax2.plot(x_date,y_search,y_osstem,marker="o")
ax2.plot(x_date,y_nft,marker="o",color='red')

ax1.set_xlabel('date')
ax1.set_xticklabels(x_date, rotation = 90)
ax1.set_ylabel('stock price')
ax2.set_ylabel('search')
plt.show()
```
