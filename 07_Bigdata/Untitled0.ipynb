```python
x_set = np.arange(2,13)
y_set = np.arange(1,7)
```

```python
def f_XY(x, y):
    if 1 <= y <= 6 and 1<= x - y <= 6:
        return y * (x-y) / 441
    else:
        return 0

print(x_set)
print(y_set)
```

```python
XY = [x_set, y_set, f_XY]
prob = np.array([[f_XY(x_i, y_j) for y_j in y_set]
                 for x_i in x_set])
print(prob)
```
